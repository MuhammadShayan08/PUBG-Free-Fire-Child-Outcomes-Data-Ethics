# 📊 Gaming & Kids: EDA on PUBG / Free Fire Impact 🚀

This notebook explores the potential impact of heavy gaming (PUBG & Free Fire) on kids' daily lives, 
focusing on screen time, health, study habits, and social interactions.  

⚠️ **Note:** This is a responsible analysis, not blaming games but understanding balance.  


## 1. Importing Libraries

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## 2. Example Synthetic Dataset

```python
# Example dataset with Muslim kids' names
data = {
    "Kid_Name": [
        "Ayaan","Zayan","Hamza","Rayyan","Daniyal","Arham","Subhan",
        "Zoraiz","Huzaifa","Shahmir","Ibrahim","Saif","Rehan","Faris",
        "Zain","Adeel","Sameer","Eesa","Salman","Umair","Talha"
    ],
    "Daily_Play_Hours": [2,3,4,5,6,1,7,3,4,2,5,6,4,7,3,5,6,2,1,4,5],
    "Study_Hours": [5,4,3,2,1,6,1,4,3,5,2,1,3,1,4,2,1,5,6,3,4],
    "Sleep_Hours": [8,7,6,6,5,8,5,7,6,8,6,5,6,5,7,6,5,8,9,6,7],
    "Aggression_Level": [3,4,6,7,8,2,9,5,6,3,7,8,6,9,4,7,8,3,2,6,5]
}

df = pd.DataFrame(data)
df.head(22)
```

## 3. Exploratory Data Analysis

```python

plt.figure(figsize=(8,5))
sns.histplot(df["Daily_Play_Hours"], bins=7, kde=True)
plt.title("Distribution of Daily PUBG/Free Fire Play Hours")
plt.show()
```

```python

plt.figure(figsize=(8,5))
sns.scatterplot(x="Daily_Play_Hours", y="Study_Hours", data=df, hue="Aggression_Level", palette="coolwarm", size="Aggression_Level", sizes=(50,200))
plt.title("Play Hours vs Study Hours (Aggression as color/size)")
plt.show()
```

## 4. Correlation Analysis

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(7,5))
sns.heatmap(numeric_df.corr(), annot=True, cmap="RdBu", center=0)
plt.title("Correlation Between Gaming, Study, Sleep & Aggression")
plt.show()
```

## 5. Insights & Conclusion


- ⏱️ Higher daily play hours often reduce study and sleep hours.  
- 😡 Aggression levels tend to increase with higher play hours.  
- ⚖️ Balance is key: gaming for fun is fine, but moderation is important.  

✅ This notebook encourages **healthy digital habits** rather than banning games.
